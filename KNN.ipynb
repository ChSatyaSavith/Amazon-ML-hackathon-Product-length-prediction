```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import tensorflow as tf
```

```python
import os

os.chdir('..') # Change to parent directory
cwd = os.getcwd() # Get current working directory
print(cwd)
```

```python
# Read the dataset using Dask
dataset = pd.read_csv('dataset/trainfinal.csv', on_bad_lines='skip', encoding='utf-8', engine='python')
# dataset =  dataset.sample(n=20000)
```

```python
dataset.columns
```

```python
dataset
```

```python
data = dataset
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python

# Split the dataset into training and testing data
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :10], data.iloc[:, 11], test_size=0.2, random_state=42)

# Scale the training and testing data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train a gradient boosting model with early stopping and reduced learning rate
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.05, random_state=42, validation_fraction=0.2, n_iter_no_change=5, tol=1e-6,verbose=1)
gb_model.fit(X_train, y_train)

# Evaluate the model on the testing data
y_pred_gb = gb_model.predict(X_test)
mse_gb = mean_squared_error(y_test, y_pred_gb)
print('MSE of Gradient Boosting:', mse_gb)
```

```python
# Train a neural network model
nn_model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])
nn_model.compile(loss='mse', optimizer='adam')
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
history = nn_model.fit(X_train, y_train, epochs=50, batch_size=64, validation_split=0.2, callbacks=[early_stop])
```

```python
# Evaluate the model on the testing data
y_pred_nn = nn_model.predict(X_test)
mse_nn = mean_squared_error(y_test, y_pred_nn)
print('MSE of Neural Network:', mse_nn)
```
