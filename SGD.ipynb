```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
# Read the dataset using Dask
dataset = pd.read_csv('dataset/trainfinal.csv', on_bad_lines='skip', encoding='utf-8', engine='python')
# dataset =  dataset.sample(n=20000)
```

```python
dataset.columns
```

```python
# dataset = dataset.drop(['Unnamed: 0'],axis=1)
```

```python
train_df = dataset
```

```python
# train_df.interpolate(method ='linear', limit_direction ='forward')
```

```python
# # Load the train dataset


# Split dataset into input features and output variable
X = train_df.iloc[:, 0:12].values
y = train_df.iloc[:, 12].values

# Convert input features and output variable into numpy arrays
# Identify and remove rows with NaN values in y
nan_indices = np.isnan(y)
X = X[~nan_indices]
y = y[~nan_indices]
```

```python
# Train the Stochastic Gradient Descent Regressor
sgd = SGDRegressor( penalty='l2', alpha=0.0001, max_iter=1000, random_state=42,verbose =1)
sgd.fit(X, y)
```

```python
import pickle
# Save the model to a file
with open('SGDfinal.pickle', 'wb') as f:
    pickle.dump(sgd, f)
```
